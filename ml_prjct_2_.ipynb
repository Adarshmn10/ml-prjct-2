# Step 6: Outlier Detection

```python
df1=df.copy()
```

```python
df1.plot(kind='box',subplots=True,layout=(4,3),figsize=(10,8),color='red')
plt.tight_layout()
plt.show()
```

```python
from scipy import stats
for i in df1.columns:
  if i !='Global_Sales':
    z_score =np.abs(stats.zscore(df1[i]))
    outliers=df1[z_score>3]
    print(f'Number of outliers in {i} : {outliers.shape[0]}')
```

```python
#removing outliers using IQR method
for i in df1.columns:
  if i !='Global_Sales':
      data=df1[i]
      Q1=data.quantile(0.25)
      Q3=data.quantile(0.75)
      IQR=Q3-Q1
      lower_bound=Q1-1.5*IQR
      upper_bound=Q3+1.5*IQR
      df1=df1[(data>=lower_bound)&(data<=upper_bound)]
```

```python
for i in df1.columns:
  if i !='Global_Sales':
    z_score =np.abs(stats.zscore(df1[i]))
    outliers=df1[z_score>3]
    print(f'Number of outliers in {i} : {outliers.shape[0]}')
```

# Step 7:feature Engineering

```python
import numpy as np
df.info()
df.skew()
# Apply log transformation to the numerical data within the DataFrame
x=['Rank','Name','Platform','Year','Genre','Publisher','NA_Sales','EU_Sales','JP_Sales','Other_Sales']
for i in x :
  df[i]=np.log1p(df[i])
df
```

**summary:-**
  * This step involves transforming existing features or creating new ones to enhance the model's performance.

```python
x=df.drop('Global_Sales',axis=1)
y=df['Global_Sales']
```

# Step 8:Data Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=scaler.fit_transform(x)
```

```python
df
```

**summary:-**
  * Data scaling is applied to standardize the features and improve the model's training proces

# Step 9: Initial Modeling

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,shuffle=True)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor(n_estimators=100)
rfr.fit(x_train,y_train)
pred=rfr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
rmse=mean_squared_error(y_test,pred)**(1/2)
print('RMSE:',rmse)
```

```python
from sklearn.metrics import r2_score
accuracy=r2_score(y_test,pred)
print('accuracy:',accuracy)
```

**summary:-**
  * A Random Forest Regressor model is trained and evaluated using metrics like RMSE and R-squared to assess its predictive accuracy.
